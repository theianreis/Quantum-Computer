```python
import numpy as np
import matplotlib.pyplot as plt

# Criando dados de exemplo (matriz 10x10)
data = np.random.rand(10, 10)  # valores aleatórios entre 0 e 1

# Criando o heatmap
plt.imshow(data, cmap='viridis', interpolation='nearest')

# Adicionando barra de cores (legenda)
plt.colorbar(label='Intensidade')

# Título e eixos
plt.title('Mapa de Calor Exemplo')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Exibir
plt.show()
```

```python
labels_x = [f'Col{i}' for i in range(10)]
labels_y = [f'Lin{i}' for i in range(10)]

plt.imshow(data, cmap='plasma', interpolation='nearest')
plt.colorbar(label='Valor')

plt.xticks(ticks=np.arange(10), labels=labels_x, rotation=45)
plt.yticks(ticks=np.arange(10), labels=labels_y)

plt.title("Mapa de Calor com Rótulos")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados (exemplo 5x5)
data = np.random.rand(5, 5)

fig, ax = plt.subplots()
im = ax.imshow(data, cmap='viridis')

# Adiciona barra de cor
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Intensidade')

# Mostra os valores dentro de cada célula
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        value = f"{data[i, j]:.2f}"  # formatação com 2 casas decimais
        ax.text(j, i, value, ha='center', va='center', color='white')

# Eixos
ax.set_title('Mapa de Calor com Valores')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Criando o DataFrame
df = pd.DataFrame({
    'A': [0.1, 0.4, 0.7],
    'B': [0.5, 0.2, 0.3],
    'C': [0.8, 0.9, 0.6]
}, index=['X', 'Y', 'Z'])

# Plot
fig, ax = plt.subplots()
im = ax.imshow(df.values, cmap='coolwarm')

# Barra de cor
plt.colorbar(im)

# Rótulos
ax.set_xticks(np.arange(len(df.columns)))
ax.set_yticks(np.arange(len(df.index)))
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.index)

# Rotacionar se quiser
plt.xticks(rotation=45)

# Inserir valores
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        valor = f"{df.iloc[i, j]:.2f}"
        cor = "white" if df.iloc[i, j] < 0.5 else "black"
        ax.text(j, i, valor, ha="center", va="center", color=cor)

plt.title("Heatmap com DataFrame")
plt.tight_layout()
plt.show()
```
