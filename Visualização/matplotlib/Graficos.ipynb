```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Gráfico de Linha
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='seno')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Gráfico de Linha')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Gráfico de Barras
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 2, 5]

plt.bar(categorias, valores, color='skyblue')
plt.title('Gráfico de Barras')
plt.ylabel('Valor')
plt.show()
```

```python
#Gráfico de Dispersão (Scatter Plot)
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green')
plt.title('Gráfico de Dispersão')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Múltiplas Linhas no Mesmo Gráfico
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sen(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='orange', linestyle='--')
plt.title('Seno e Cosseno')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Linha com Gradiente de Cor (colormap com cor ao longo da linha)

from matplotlib.collections import LineCollection

x = np.linspace(0, 10, 100)
y = np.sin(x)
points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

lc = LineCollection(segments, cmap='plasma', linewidth=2)
lc.set_array(x)

fig, ax = plt.subplots()
ax.add_collection(lc)
ax.autoscale()
ax.set_title('Linha com Gradiente de Cor')
plt.colorbar(lc, ax=ax, label='x')
plt.show()
```

```python
#Gráfico com Marcadores (Markers)
x = np.linspace(0, 10, 10)
y = np.log(x + 1)

plt.plot(x, y, marker='o', linestyle='-', color='green')
plt.title('Gráfico com Marcadores')
plt.xlabel('x')
plt.ylabel('log(x+1)')
plt.grid(True)
plt.show()
```

```python
#Gráfico com Linha de Erro (Error Bar)

x = np.linspace(0, 10, 10)
y = np.sin(x)
erro = 0.1 + 0.1 * np.sqrt(x)

plt.errorbar(x, y, yerr=erro, fmt='-o', capsize=5)
plt.title('Gráfico com Barra de Erro')
plt.xlabel('x')
plt.ylabel('sen(x)')
plt.grid(True)
plt.show()
```

```python
#Gráfico de Linha com Estilo Científico (LaTeX e eixos log)
x = np.linspace(0.1, 10, 100)
y = np.exp(x)

plt.plot(x, y, label=r'$e^x$', color='purple')
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$x$ (escala log)')
plt.ylabel(r'$y = e^x$ (escala log)')
plt.title('Gráfico com Escalas Logarítmicas')
plt.legend()
plt.grid(True, which='both', linestyle='--')
plt.show()
```
