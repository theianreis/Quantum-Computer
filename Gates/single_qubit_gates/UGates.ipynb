- A UGate é a forma mais geral de representar uma rotação arbitrária de um qubit puro.

$$U(\theta, \phi, \lambda) = R_z(\phi) \cdot R_y(\theta) \cdot R_z(\lambda)$$

- $( R_z(\lambda))$: rotação em torno do eixo \( z \)
- $( R_y(\theta))$: rotação em torno do eixo \( y \)
- $( R_z(\phi))$: rotação final em torno do eixo \( z \)

- Esses parâmetros permitem representar qualquer rotação unitária em um qubit, o que torna essa porta muito poderosa e flexível.

$$
U(\theta, \phi, \lambda) =
\begin{bmatrix}
\cos\left(\frac{\theta}{2}\right) & -e^{i\lambda} \sin\left(\frac{\theta}{2}\right) \\
e^{i\phi} \sin\left(\frac{\theta}{2}\right) & e^{i(\phi + \lambda)} \cos\left(\frac{\theta}{2}\right)
\end{bmatrix}
$$

```python
from qiskit import *
from qiskit_aer import *
import random
import numpy as np
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector

# Geração de parâmetros aleatórios
theta = random.uniform(0, 2 * np.pi)
phi = random.uniform(0, 2 * np.pi)
lam = random.uniform(0, 2 * np.pi)

# Criação do circuito com 1 qubit e 1 bit clássico
qc = QuantumCircuit(1, 1)

qc.draw('mpl')

```

```python
state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
# Aplica a UGate usando o método direto
qc.u(theta, phi, lam, 0)

# Mede o qubit
#qc.measure(0, 0)

# Mostra o circuito
qc.draw('mpl')
```
